```python
def bag_colours(instructions = list):
    bag = ["shiny gold bag"]
    total = set()
    tmp = []
    while bag:
        for b in bag:
            pattern = re.compile(rf"(.*)s contain.*({b})")
            for i in instructions:
                res = pattern.search(i)
                if res:
                    tmp.append(res.group(1))
        total.update(tmp)
        bag = tmp
        tmp = []
    return len(total)


    
```

Part 2 - putting data into dictionary

```python
bag_dict = {}
with open("input.txt", "r") as f:
    for l in f:
        key, inside = l.strip().split(" bags contain ")
        tmp = inside.split(", ")
        tmp_dict = {}
        for t in tmp:
            if t.startswith("no"):
                break
            else:
                keys = re.search("(\d+) (.*) bag", t)
                tmp_dict[keys.group(2)] = int(keys.group(1))
        bag_dict[key] = tmp_dict
            
            
    
```

```python
def search(bag, b_dict = bag_dict):
    count = 0
    b = b_dict[bag]
    for k in b:
        v = b[k]
        count += v + v * search(k)
    return count
    
    
```
