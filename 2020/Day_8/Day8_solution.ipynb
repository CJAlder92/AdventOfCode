```python
with open("input.txt", "r") as f:
    code_list = [l.strip().split() for l in f]
```

```python
def run_program(codes):
    run_code = []
    i = 0
    acc = 0
    while i not in run_code:
        run_code.append(i)
        if codes[i][0] == "nop":
            i += 1
        if codes[i][0] == "acc":
            acc += int(codes[i][1])
            i += 1
        if codes[i][0] == "jmp":
            i += int(codes[i][1])
        if i > len(codes):
            return f"Completed, value is {acc}"        
    return f"Repeated, value is {acc}"
```

```python
run_program(code_list)
```
