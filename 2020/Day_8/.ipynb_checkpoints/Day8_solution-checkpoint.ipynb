```python
with open("input.txt", "r") as f:
    code_list = [l.strip().split() for l in f]
```

```python
def run_program(codes):
    run_code = []
    i = 0
    acc = 0
    while i not in run_code:
        run_code.append(i)
        if codes[i][0] == "nop":
            i += 1
        if codes[i][0] == "acc":
            acc += int(codes[i][1])
            i += 1
        if codes[i][0] == "jmp":
            i += int(codes[i][1])
        if i >= len(codes):
            return f"Completed, value is {acc}"
    return f"Repeated , value is {acc}"
```

```python
run_program(code_list)
```

Part 2

```python
indices = [i for i, x in enumerate(code_list) if x[0] in ["jmp", "nop"]]
indices
```

```python
nested_codes = []
for i in indices:
    tmp = [x[:] for x in code_list]
    if tmp[i][0] == "jmp":
        tmp[i][0] = "nop"
    if tmp[i][0] == "nop":
        tmp[i][0] = "jmp"
    nested_codes.append(tmp)
```

```python
res = []
for code in nested_codes:
    res.append(run_program(code))
```

```python
len(nested_codes)
```

```python
for r in res:
    if r.startswith("Completed"):
        print(r)
```
